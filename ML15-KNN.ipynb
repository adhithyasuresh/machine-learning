```python
from sklearn.datasets import load_iris
```

```python
i=load_iris()
i
```

```python
i.feature_names
```

```python
i.target_names
```

```python
i.target
```

```python
import pandas as pd
a=pd.DataFrame(i.data,columns=i.feature_names)
a
```

```python
a["column_names"]=i.target
a
```

```python
a0=a[:50]
a0
```

```python
a1=a[50:100]
a1
```

```python
a2=a[100:]
a2
```

```python
import matplotlib.pyplot as plt
plt.scatter(a0["sepal length (cm)"],a0["sepal width (cm)"],color="red",marker="*")
plt.scatter(a1["sepal length (cm)"],a1["sepal width (cm)"],color="black",marker=".")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
x=a.drop("column_names",axis=1)
y=a["column_names"]
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=30)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=10)               #default aayitt n_neighbors=5 aayirikkum
model.fit(x_train,y_train)
```

```python
y_pred_train=model.predict(x_train)
y_pred_train
```

```python
y_pred_test=model.predict(x_test)
y_pred_test
```

```python
from sklearn.metrics import accuracy_score
a_s=accuracy_score(y_test,y_pred_test)
a_s
```

```python
a_s=accuracy_score(y_train,y_pred_train)
a_s
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
classification_report(y_test,y_pred_test)
```

```python
cm=confusion_matrix(y_test,y_pred_test)
cm
```

```python
import seaborn as sns
sns.heatmap(cm,annot=True)
```
